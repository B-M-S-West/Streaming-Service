```python
data_year = data.copy()
data_year = data_year.dropna(subset=['date_added']) # removes the couple of blank rows
data_year['date_added'] = data_year['date_added'].str.strip() # removes any spaces
data_year['date_added'] = pd.to_datetime(data_year['date_added'], format = '%B %d, %Y')
data_year['year'] = data_year['date_added'].dt.strftime('%Y')
data_year = data_year.sort_values(by='date_added')
data_year.head()
```

```python
# group by type and the month_year, and calculate the cumulative sum
year_cumulative = data_year.groupby(['type', 'date_added']).size().groupby(level=[0]).cumsum().reset_index(name='count')

# create a line chart showing the cumulative amount of movies and TV shows added
fig, ax = plt.subplots(figsize=(10, 6))
for key, grp in year_cumulative.groupby(['type']):
    ax = grp.plot(ax=ax, kind='line', x='date_added', y='count', label=key)
plt.title('Cumulative amount of movies and TV shows added')
plt.xlabel('Date')
plt.ylabel('Count')
plt.show()
```

```python
# create the years to release column
data_year['years_to_release'] = (data_year['year'].astype(int))-(data_year['release_year'].astype(int))

# plot the bar chart of the frequency of each unique value
data_year['years_to_release'].plot.hist(bins=10, range=(0, 10))

# set title and labels
plt.title('How long it took films to be released on Netflix')
plt.xticks(range(0, 11, 1))
plt.xlabel('Years to release')
plt.ylabel('Amount')

plt.show()
```

```python
# Split the listed_in column into individual classifications
classification = data['listed_in'].str.split(',', expand = True)
classification = classification.applymap(lambda x: x.lstrip() if isinstance(x, str) else x)

# Count the frequency of each classification
classification_count = classification.stack().value_counts()

# Plot the different classification types as a bar chart
plt.figure(figsize = (10, 6))
sns.barplot(x = classification_count.index, y = classification_count.values)
plt.xlabel('Classification')
plt.ylabel('Frequency')
plt.title('Frequency of different classifications on Netflix')
plt.xticks(rotation = 90)
plt.show
```

```python
# Create a copy of the data for just movies
film = data[data['type'] == 'Movie']

# Split the listed_in column into individual classifications
film_classification = film['listed_in'].str.split(',', expand = True)
film_classification = film_classification.applymap(lambda x: x.lstrip() if isinstance(x, str) else x)

#Combine into a series
film_classification = film_classification.stack()

# Count the frequency of each classification, but drop 'Movies'
film_classification_count = film_classification.drop(film_classification[film_classification == 'Movies'].index).value_counts()


# Plot the different classification types as a bar chart
plt.figure(figsize = (10, 6))
sns.barplot(x = film_classification_count.index, y = film_classification_count.values)
plt.xlabel('Classification')
plt.ylabel('Frequency')
plt.title('Frequency of different Movie classifications on Netflix')
plt.xticks(rotation = 90)
plt.show
```

```python
# Create a copy of the data for just TV Shows
tv_show = data[data['type'] == 'TV Show']

# Split the listed_in column into individual classifications
tv_show_classification = tv_show['listed_in'].str.split(',', expand = True)
tv_show_classification = tv_show_classification.applymap(lambda x: x.lstrip() if isinstance(x, str) else x)

#Combine into a series
tv_show_classification = tv_show_classification.stack()

# Count the frequency of each classification, but drop 'TV Show'
tv_show_classification_count = tv_show_classification.drop(tv_show_classification[tv_show_classification == 'TV Shows'].index).value_counts()

# Plot the different classification types as a bar chart
plt.figure(figsize = (10, 6))
sns.barplot(x = tv_show_classification_count.index, y = tv_show_classification_count.values)
plt.xlabel('Classification')
plt.ylabel('Frequency')
plt.title('Frequency of different TV Show classifications on Netflix')
plt.xticks(rotation = 90)
plt.show
```

```python
def my_rating_func(dataframe):
    
    dataframe['rating_group']= dataframe['rating'].astype(str)
    
    for index, row in dataframe.iterrows():
        if row['rating'] == "TV-MA" or row['rating'] == "R": 
            dataframe.at[index, 'rating_group'] = "For Mature Audiences"
        elif row['rating'] == "TV-14" or row['rating'] == "PG-13":
            dataframe.at[index, 'rating_group'] = "PG 13+"
        elif row['rating'] == "NR":
            dataframe.at[index, 'rating_group'] = "Not Rated"
        else:
            dataframe.at[index, 'rating_group'] = "PG 13-"
            
    return dataframe
```

```python
ratings = my_rating_func(data)
```

```python
rg_chart = ratings.rating_group.value_counts()
rg_chart
```

```python
chart = rg_chart.plot(kind='bar')
```

```python
rg_chartsum = ratings.rating_group.value_counts().sum()
rg_chartsum
```

```python
rg_chart_pct = round((ratings.rating_group.value_counts() / rg_chartsum)*100)
rg_chart_pct
```

```python
chart = rg_chart.plot(kind='pie')
```

```python
rg_chart2 = (ratings.groupby('type')['rating_group'].value_counts()).unstack()
rg_chart2
```

```python
chart2 = rg_chart2.plot(stacked=True, kind='bar')
```

```python
rg_chart2 = (ratings.groupby('type')['rating_group'].value_counts()).unstack()
rg_chart2
```

```python
rg_chart2['Total'] = rg_chart2[['For Mature Audiences', 'Not Rated', 'PG 13+', 'PG 13-']].sum(axis=1)
rg_chart2
```

```python
rg_chart3 = rg_chart2.copy()
rg_chart3['For Mature Audiences'] = round((rg_chart3['For Mature Audiences'] / rg_chart3['Total'])*100)
rg_chart3['Not Rated'] = round((rg_chart3['Not Rated'] / rg_chart3['Total'])*100)
rg_chart3['PG 13+'] = round((rg_chart3['PG 13+'] / rg_chart3['Total'])*100)
rg_chart3['PG 13-'] = round((rg_chart3['PG 13-'] / rg_chart3['Total'])*100)
rg_chart3 = rg_chart3.drop(['Total'],axis=1)
rg_chart3
```

```python
chart3 = rg_chart3.plot(stacked=True, kind='bar')
```
